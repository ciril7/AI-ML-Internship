```python
import pandas as pd
import seaborn as sns
import sklearn.preprocessing
```

```python
df=pd.read_csv('/content/wine_quality_merged.csv')
```

```python
df.info()
```

```python
df['pH_mean_imputed']=df['pH'].fillna(df['pH'])
df['pH_median_imputed']=df['pH'].fillna(df['pH'].median())
df['alcohol_mode_imputed']=df['alcohol'].fillna(df['alcohol'].mode()[0])
df[['pH','pH_mean_imputed','pH_median_imputed']]
```

```python
numerical_df = df.select_dtypes(include=['float64', 'int64'])

corr_matrix = numerical_df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Wine Quality Features')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
numerical_columns=['alcohol','density']
```

```python
df.dropna(subset=numerical_columns,inplace=True)
```

```python
minmax_scaler=MinMaxScaler()
standard_scaler=StandardScaler()
```

```python
df['alcohol_MinMax']=minmax_scaler.fit_transform(df[['alcohol']])
df['alcohol_Standard']=standard_scaler.fit_transform(df[['alcohol']])


df['density_MinMax']=minmax_scaler.fit_transform(df[['density']])
df['density_Standard']=standard_scaler.fit_transform(df[['density']])
```

```python
df[['alcohol','alcohol_MinMax','alcohol_Standard']]
df[['density','density_MinMax','density_Standard']]
```
