```python
import pandas as pd

```

```python
df=pd.read_csv("/content/titanic uncleaned.csv")
```

```python
df.info()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['Age_mean_imputed']=df['Age'].fillna(df['Age'])
```

```python
df['Age_median_imputed']=df['Age'].fillna(df['Age'].median())
```

```python
df['Cabin_mode_imputed']=df['Cabin'].fillna(df['Cabin'].mode()[0])
```

```python
df[['Age','Age_mean_imputed','Age_median_imputed']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Pclass',y='Age',palette='pastel')
plt.title('Boxplot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x='Pclass',y='Survived',palette='pastel')
plt.title('Boxplot of Survived Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survived')
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Pclass',y='Age',palette='pastel')
plt.title('bar
plot of Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
```
